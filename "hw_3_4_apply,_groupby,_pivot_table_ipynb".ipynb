2. В якому місті знайходиться філіал з найвищим прибутком? Може в тому місці нам розмітисти ще один магазин.  
Знайдіть відповідь за допомогою функціоналу Pandas.

```python
city_profit = df.groupby('City')['gross income'].sum()
city_profit
```

```python
# Знаходимо місто з найвищим прибутком
max_profit_city = city_profit.idxmax()

print(f"Філіал з найвищим прибутком розташований в місті: {max_profit_city}")
```

3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.

3. Які товари користуються найменьшим попитом? Яка дохідність від них?

```python
demand = df.groupby('Product line')[['Quantity', 'Total']].sum().reset_index()
demand
```

```python
min_quantity_index = demand['Quantity'].idxmin()
min_quantity_index
```

```python
min_quantity_item = demand.loc[min_quantity_index, 'Product line']
min_quantity_item
```

```python
min_quantity = demand.loc[min_quantity_index, 'Quantity']
min_quantity
```

```python
print(f'Товари {min_quantity_item} користуються найнижчим попитом у кількості {min_quantity} покупок, але кількість покупок меньше від покупок інших товарів не кретично, потрібно ще додатково аналізувати можливо дохідність та маржу')
```
